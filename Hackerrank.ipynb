```python
n = int(input())
L = []
for _ in range(n):
    commands = input().split()
    command = commands[0]
    args = commands[1:]
    if command != "print":
        if command == "insert":
            L.insert(int(args[0]), int(args[1]))
        elif command=="remove":
            L.remove(int(args[0]))
        elif command=="append":
             L.append(int(args[0]))
        elif command=="sort":
             L.sort()
        elif command=="pop":
             L.pop()
        elif command=="reverse":
             L.reverse()
    else:
        print(L)
    
```

```python
def count_substring(string, sub_string):
    i = string.find(sub_string)
    indices = []
    while i >= 0:
        indices.append(i)
        i = string.find(sub_string, i+1)
    return len(indices)

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```
